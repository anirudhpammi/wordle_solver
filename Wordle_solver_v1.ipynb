# Load dictionary

```python
direc = 'words_dictionary.json'
with open(direc, 'r') as f:
  data = json.load(f)
```

# Find all five letter words

```python
# list(data.keys())[-1] last element of dict
five_letter = []
for i in data:
    if(len(i)==5):
        five_letter.append(i)
print("Total 5 letter words: ",len(five_letter))
```

```python
five_letter
```

# Generate a state vector for all the words

```python
lett_mat = np.zeros((len(five_letter),26))

cnt = 0
for i in five_letter:
    for j in i:
        #print(j,ord(j)-97,cnt)
        lett_mat[cnt,ord(j)-97] += 1
    cnt+=1
    
[state,count] = np.unique(lett_mat,return_counts=True,axis=0)
count_dres_arg = np.argsort(count)[::-1]
state_des = state[count_dres_arg] #states in descending order of prob
print("Max count: ",count[count_dres_arg[0]])
print("Corresponding state ",state[count_dres_arg[0]])
```

# Analysis of "orthogonality" of words with decreasing probability

```python
a = []
cnt = 0
for i in lett_mat:
    if(np.all(i==state_des[0])):
        a.append(cnt)
    cnt+=1
print("Printing all the words corresponding to the most probable state")
for k in a:
    print(five_letter[k])
```

```python
dot_states = (np.dot(state_des,state_des[0,:].reshape(26,1))).reshape(state.shape[0])

cnt0 = 0
cnt1 = 0
cnt2 = 0
for i in range(state.shape[0]):
    if (dot_states[i] == 0 and cnt0==0):
        print("Closest index with dot product 0 = ", i)
        cnt0+=1
    if (dot_states[i] == 1 and cnt1==0):
        print("Closest index with dot product 1 = ", i)
        cnt1+=1

    if (dot_states[i] == 2 and cnt2==0):
        print("Closest index with dot product 2 = ", i)
        cnt2+=1
```

```python
def print_words(five_letters,state_des,dot_val,index):
    a = []
    cnt = 0
    for i in lett_mat:
        if(np.all(i==state_des[index])):
            a.append(cnt)
        cnt+=1
    print("Printing all the words corresponding dot prod =",dot_val)
    for k in a:
        print(five_letter[k])
        
        
print_words(five_letter,state_des,2,5)
print_words(five_letter,state_des,1,42)
print_words(five_letter,state_des,0,394)
```

# Guessing

```python
word = np.zeros(5) - 1
possible_letters = ['e','s','o','h','t']


#if position is known enter here
#word[0]= ord('t')-97
#word[1]= ord('h')-97
#word[2]= ord('')-97
#word[3]= ord('')-97
#word[4]= ord('')-97


known_lett_num = np.sum(word>-1)
tem_vec = np.zeros(known_lett_num)

len_letters = len(possible_letters)
possible_letters_vec = np.zeros(len_letters)

for i in range(len_letters):
    possible_letters_vec[i] = ord(possible_letters[i])-97

for i in range(len_letters-1):
    if (i==0):
        ls1 = (np.where(state_des[:,int(possible_letters_vec[i])] > 0)[0])
    
    ls2 = (np.where(state_des[:,int(possible_letters_vec[i+1])] > 0)[0])
    ls1 = np.intersect1d(ls1,ls2)
```

```python
print("Printing all the words possible with the guess with positions")
for jj in range(ls1.shape[0]):
    a = []
    cnt = 0
    for i in lett_mat:
        if(np.all(i==state_des[ls1[jj]])):
            a.append(cnt)
        cnt+=1
    print("\n Anagram break, (1 state)")
    for k in a:
        if (np.sum(word)==-5):
            print(five_letter[k])
        else:
            known_letters = np.where(word>-1)[0]
            
            tem_vec = np.zeros(known_lett_num)
            for ii in range(known_lett_num):
                tem_vec[ii] = (ord(five_letter[k][known_letters[ii]])-97)
                if(np.all(word[known_letters] == tem_vec)):
                    print(five_letter[k])
                
        
```
